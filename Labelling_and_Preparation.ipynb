```python
import pandas as pd
```

```python
df_crawl = pd.read_csv('dataset_review_tokped.csv', index_col=[0])
```

```python
df_crawl
```

```python
#remove messy enter charracter
df_crawl.replace(to_replace=[r"\\t|\\n|\\r", "\t|\n|\r"], value=[" "," "], regex=True, inplace=True)
```

```python
df_crawl
```

```python
import re
```

```python
#remove emoji
df_crawl['Review'] = df_crawl['Review'].str.replace('[^\w\s#@/:%.,_-]', '', flags=re.UNICODE)
# df = df.astype(str).apply(lambda x: x.str.encode('ascii', 'ignore').str.decode('ascii'))
df_crawl
```

```python
#conver data to lowercase
df_crawl['Review'] = df_crawl['Review'].str.lower()
df_crawl
```

```python
# remove punctuation
df_crawl["Review"] = df_crawl['Review'].str.replace('[^\w\s]',' ')
df_crawl
```

```python
# multiple space
df_crawl['Review'] = df_crawl['Review'].str.replace(' {2,}', ' ', regex=True)
df_crawl['Review'] = df_crawl['Review'].str.strip()
df_crawl
```

```python
#Drop Empty Rows
print(df_crawl.isnull().sum())
df_crawl.dropna(inplace = True)
print(df_crawl.isnull().sum())
df_crawl['Review'].astype(bool)
df_crawl = df_crawl[df_crawl['Review'].astype(bool)]
df_crawl
```

```python
#Removes duplicates in place
df_crawl.drop_duplicates(keep=False, inplace=True)
df_crawl
```

```python
# saving the dataset
df_crawl.to_csv('dataset_review_tokped_cleaned.csv', index=False)
df_crawl
```
